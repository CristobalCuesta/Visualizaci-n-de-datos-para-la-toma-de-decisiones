```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mlp
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_excel('datasets/mic.xlsx',
                   engine = 'openpyxl'
                  )
df.head()
```

```python
counts = df['Age'].value_counts()
type(counts)
counts = counts.to_frame()
type(counts)
counts
counts['%'] = (counts.iloc[:,:]/counts.loc[:, 'Age'].sum(axis = 0))*100
counts
```

```python
counts = counts.reindex(['<=20', '21 - 30', '31 - 40', '41 - 50'])
counts.reset_index(inplace = True)
counts
```

```python
counts = counts.rename({'index':'Age_group'}, axis = 1
                      )
counts
```

```python
explode_list = [0.04,0.04,0.04,0.04] #separación de los trangulitos
clor_list = ['gold', 'yellowgreen', 'coral', 'skyblue']
counts['Age'].plot(kind = 'pie',
                  figsize = (15, 6),
                   startangle = 90,
                   shadow = True,
                   labels = None,
                   autopct = '%1.1f%%',
                   pctdistance = 1.1,
                   explode = explode_list,
                   colors = clor_list,
                  )
plt.legend(labels = counts.Age_group,
           bbox_to_anchor = (0.9, 0.9)
          )
plt.title('Numero de respuestas de diferentes grupos de edad', fontsize = 17, y = 1)
plt.show()
```

```python
x = list(counts['Age_group'])
y = list(counts['Age'])

plt.bar(x, y, color = 'darkcyan')
plt.title('Numero de respuestas de diferentes grupos de edad', fontsize = 17, y = 1)

plt.show()
```

```python
plt.subplot(1,2,1)

explode_list = [0.04,0.04,0.04,0.04] #separación de los trangulitos
clor_list = ['gold', 'yellowgreen', 'coral', 'skyblue']
counts['Age'].plot(kind = 'pie',
                  figsize = (15, 6),
                   startangle = 90,
                   shadow = True,
                   labels = None,
                   autopct = '%1.1f%%',
                   pctdistance = 1.1,
                   explode = explode_list,
                   colors = clor_list,
                  )
plt.legend(labels = counts.Age_group,
           bbox_to_anchor = (0.9, 0.9)
          )
plt.title('Numero de respuestas de diferentes grupos de edad', fontsize = 17, y = 1)

plt.subplot(1,2,2)

x = list(counts['Age_group'])
y = list(counts['Age'])

plt.bar(x, y, color = 'darkcyan')
plt.title('Numero de respuestas de diferentes grupos de edad', fontsize = 17, y = 1)

plt.show()
```

```python
age_genro = df.groupby(['Age', 'Sex']).size()
type(age_genro)
age_genero = age_genro.to_frame()
age_genero
```

```python
age_genero_pivot = pd.pivot_table(index =['Age', 'Sex'],
                                  aggfunc = len,
                                  data = df,
                                 values = 'Availability')
age_genero_pivot
```

```python
age_gender = age_genero_pivot.rename(columns = {'Availability': 'Count'}
                                    )
age_gender['%'] = (age_gender.iloc[:,:] / age_gender.iloc[:,:].sum(axis = 0)) * 100
age_gender
```
