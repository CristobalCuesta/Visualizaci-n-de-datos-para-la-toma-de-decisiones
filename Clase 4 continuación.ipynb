```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/tips.csv')
#df.head()
```

```python
sns.distplot(df['total_bill'], #histplot y distplot es lo mismo
             #kde=False  #Poner este para quital la linea de tendencia 
             bins = 20  # aumentar el numero de varibles es la grafica 
            ).set_title('Histograma de total bill', fontsize =16)
plt.show()
```

```python
sns.boxplot(x= df['total_bill'],
            y = df['day'],
            whis = (0, 100),
           ).set_title('Box plot of total_bill')
plt.show()
```
