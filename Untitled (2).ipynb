```python
df = pd.read_csv('datasets/census.csv')
df.head(2)
```

```python
my_pivot = pd.pivot_table(df,
                          index = 'STNAME',
                          values = 'BIRTHS2010',
                          aggfunc = ['mean','sum','median'],
                          columns = 'REGION',
                          margins = True
                         )
my_pivot
```

```python
df =pd.read_csv('datasets/airline_data.csv',
               low_memory=False
               )
df.head()
```

```python
a = df.columns.tolist()
a.sort()
a
```

```python
dep_delay_pivot = pd.pivot_table (df,
                                 index= ['Year',
                                  'Quarter'],
                                  values = 'DepDelay',
                                  columns = 'DepartureDelayGroups',
                                  aggfunc = 'count',
                                 )
dep_delay_pivot
```

```python
df.groupby(['DepDelay','DepartureDelayGroups']).size()
```
