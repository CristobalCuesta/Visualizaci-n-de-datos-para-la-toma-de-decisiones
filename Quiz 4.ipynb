```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
df
```

```python
%ls datasets/
```

```python
df.describe()
df.columns
```

```python
sns.boxplot(x= df['talla'],
            whis = (0, 100),
           ).set_title('Box plot')
plt.show()
```

```python
tabla = pd.pivot_table(df,
                          index = ['desc_ent'],
                          columns = ['sexo'],
                          values = ['peso'],
                          aggfunc = {'peso':['median']}
                         )
tabla
```

```python
tabla.columns
```

```python
boxplot = tabla.boxplot(column=[('peso', 'median', 2)])
```

```python
tabla2 = pd.pivot_table(df,
                          index = ['desc_ent'],
                          columns = ['sexo'],
                          values = ['cintura'],
                          aggfunc = {'cintura':['median']}
                         )
tabla2
```

```python
lineplot = tabla2[('cintura', 'median', 1)].plot(kind='line')
```

```python
tabla3 = pd.pivot_table(df,
                          index = ['desc_ent'],
                          columns = ['sexo'],
                          values = ['sistol'],
                          aggfunc = {'sistol':['sum']}
                         )
tabla2
```

```python
by_ropa = tabla3.sort_values(('sistol', 'sum', 2), ascending= False)
top10 = by_ropa.head(10)
top10
```

```python
top10[('sistol', 'sum', 2)].plot(kind = 'barh',
                               figsize = (20, 10),
                               color = 'y',
                               alpha = 0.5)
```
