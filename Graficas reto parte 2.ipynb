```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%ls datasets/
```

```python
df = pd.read_excel('Graficas_reto_2.xlsx',
                   sheet_name='G2',
                   engine = 'openpyxl'
                  )
df.head()
```

```python
x = list(df['Indusrias diversas'])
y = list(df['Suma de Venta SKU'])

plt.bar(x, y, color = 'green')
plt.title('Venta de industrias diversas', fontsize = 17, y = 1)

plt.xticks(fontsize = 10, rotation =90)
plt.yticks(fontsize = 10)
plt.show()
```
