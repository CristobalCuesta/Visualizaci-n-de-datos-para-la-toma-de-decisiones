## Si exsiste alguna correlaciÃ³n 'Wispang' y 'CWDistance'

```python
df = pd.read_csv('datasets/Cartwheeldata.csv')
df.head()
```

```python
import seaborn as sns
```

```python
##pip install seaborn
```

```python
sns.lmplot(x = 'Wingspan',
           y = 'CWDistance',
           data = df,
           fit_reg = False, # eliminar la linea de regression
           hue = 'Gender', # para ver colores por cada caterogira
          )
plt.title('Scatter plot of Wingspan vs CWDistance', fontsize = 17)
```

```python
sns.swarmplot(x = 'Gender',
              y = 'CWDistance',
              data = df
             )
plt.title('Swarm plot of Gender vs CWDistance', fontsize = 16)
plt.show()
```

```python
sns.boxplot(data =df.loc[df['Gender'] == 'F', 
                         ['Age', 'Height', 'Wingspan', 'CWDistance']
                        ]
           )

plt.xticks(fontsize = 17, rotation = 45)
plt.yticks(fontsize = 17)
plt.title('Boxplots', fontsize =17)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1) #2 renglon, 2 columnas, 1st position
sns.boxplot(data = df.loc[df['Gender'] == 'F',
                          ['Wingspan']
                         ]
           )

ax2 = fig.add_subplot(2,2,3) #2 renglon, 2 columnas, position 3
sns.boxplot(data = df.loc[df['Gender'] == 'M',
                          ['Wingspan']
                         ]
           )

ax3 = fig.add_subplot(2,2,2) #2 renglon, 2 columnas, 2nd position
sns.boxplot(data = df.loc[df['Gender'] == 'F',
                          ['Wingspan']
                         ]
           )

ax4 = fig.add_subplot(2,2,4) #2 renglon, 2 columnas, position 4
sns.boxplot(data = df.loc[df['Gender'] == 'M',
                          ['Wingspan']
                         ]
           )

ax1.title.set_text('Genero F Wingspan')
ax2.title.set_text('Genero M Wingspan')
```
